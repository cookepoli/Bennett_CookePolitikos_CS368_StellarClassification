```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
x = np.linspace(0,1,num=100)
entropy = []
gini = []
l_loss_t = []
l_loss_f = []
l_l = []
for i in x:
    try:
        entropy.append(-((i)*math.log2(i)+(1-i)*math.log2(1-i)))
    except:
        entropy.append(0)
    gini.append(1-((i)**2+(1-i)**2))
    try:
        llt = -math.log(i)
        l_loss_t.append(llt)
    except:
        llt = 0
        l_loss_t.append(5)
    try:
        llf = -math.log(1-i)
        l_loss_f.append(llf)
    except:
        llf = 0
        l_loss_f.append(5)
        
    l_l.append(llt+llf)
    
```

```python
plt.plot(x,entropy)
plt.plot(x,gini)
#plt.plot(x,l_loss_t)
#plt.plot(x,l_loss_f)
#plt.plot(x,l_l)
plt.legend(['Entropy','Gini'])
plt.xlabel('True P(t)')
plt.ylabel('Purity')
plt.title('Purity Measures for a Binary Classification Problem')

plt.show
```
